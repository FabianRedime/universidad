```python
import gensim
from gensim.utils import simple_preprocess

def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data = data2019.description_processed.values.tolist()
data_words = list(sent_to_words(data))

print(data_words[:1][0][:30])
```

```python
#Construir los modelos de bigrama y trigrama
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Manera más rápida de convertir una oración en un trigrama/bigrama
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python

# NLTK Stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

# Definir funciones para palabras vacías, bigramas, trigramas y lematización
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out



print(data_words[:1][0][:30])
```

```python
import spacy

# Eliminar palabras vacías
data_words_nostops = remove_stopwords(data_words)

#Formar bigramas
data_words_bigrams = make_bigrams(data_words_nostops)

# Inicialice el modelo spacy 'en', manteniendo solo el componente etiquetador (para mayor eficiencia)
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])

# Haga la lematización manteniendo solo sustantivo, adj, vb, adv
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

print(data_lemmatized[:1][0][:30])
```

```python
#Transformación de Datos: Corpus y Diccionario
import gensim.corpora as corpora
id2word = corpora.Dictionary(data_words)

texts= data_words

corpus = [id2word.doc2bow(text) for text in texts]

print(corpus[:1][0][:30])
```

```python
#Imprimimos los topicos
import gensim
from gensim import corpora

# creamos el diccionario
dictionary = corpora.Dictionary(data_words)

# creamos el corpus
corpusMore = [dictionary.doc2bow(doc) for doc in data_words]

# Entrenar el modelo LDA
num_topics = 5  # Especifique el número deseado de temas
lda_model = gensim.models.LdaModel(corpusMore, num_topics=num_topics, id2word=dictionary)

# Obtenga las palabras principales para cada tema
top_words_per_topic = []
num_words = 10  # Especifique el número de palabras principales para recuperar

for topic_id in range(num_topics):
    topic_words = lda_model.get_topic_terms(topic_id, topn=num_words)
    top_words = [dictionary[word_id] for word_id, _ in topic_words]
    top_words_per_topic.append(top_words)

# Mostrar las palabras principales de cada tema
for topic_id, top_words in enumerate(top_words_per_topic):
    print(f"Topic {topic_id + 1}:")
    print(", ".join(top_words))
    print()
```

```python
#el primer topico por cada ducumento
import gensim
from gensim import corpora

# Crear el diccionario de palabras
diccionario = corpora.Dictionary(data_words)

# Crear el corpus
corpusRelevante = [diccionario.doc2bow(doc) for doc in data_words]

# Entrenar el modelo LDA
num_topics = 5  # Especifica el número de tópicos deseado
modelo_lda = gensim.models.LdaModel(corpusRelevante, num_topics=num_topics, id2word=diccionario)

# Obtener el tópico más relevante para cada documento
for i, doc in enumerate(corpus):
    topicos = modelo_lda.get_document_topics(doc, minimum_probability=0.0)
    topicos_ordenados = sorted(topicos, key=lambda x: x[1], reverse=True)
    tópico_más_relevante = topicos_ordenados[0][0]
     # Obtener las palabras clave del tópico más relevante
    palabras_clave = modelo_lda.show_topic(tópico_más_relevante, topn=10)
    palabras_clave = [palabra for palabra, _ in palabras_clave]
    print(f"Documento {i+1}: Tópico {tópico_más_relevante}, Palabras clave: {', '.join(palabras_clave)}")
```

```python
# Construir modelo LDA
from pprint import pprint
num_topics=100
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics, 
                                       random_state=100,
                                       chunksize=100,
                                       passes=10,
                                       per_word_topics=True)
pprint(lda_model.print_topic(2))
doc_lda = lda_model[corpus]
```

```python
# Generar la matriz LDA de tópicos
matriz_lda = lda_model.show_topics(formatted=False)

# Imprimir la matriz LDA de tópicos
for i, topic in matriz_lda:
    print(f'Tópico {i}: {topic}')
```

```python
from gensim.models import CoherenceModel

# Calcule la puntuación de coherencia
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```

```python

def compute_coherence_values(corpus, dictionary, k, a, b):
    
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=100,
                                           chunksize=80,
                                           passes=10,
                                           alpha=a,
                                           eta=b)
    
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words, dictionary=id2word, coherence='c_v')
    
    return coherence_model_lda.get_coherence()
```

```python
# Leemos todos los elementos del archivo 2019
lecturaCsv=pd.read_csv("dataset2019.csv")
contador_filasCsv=len(lecturaCsv)
data2019 = pd.read_csv("dataset2019.csv").sample(contador_filasCsv)
data2019.head()
```

```python
# Cargamos la libreria RE 
import re

data2019['description_processed'] = \
data2019['description'].map(lambda x:re.sub('[,\.!?]', '', x))
data2019['description_processed'] = \
data2019['description_processed'].map(lambda x: x.lower())
data2019['description_processed'].head()
```

```python
import gensim
from gensim.utils import simple_preprocess
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
            
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc))
            if word not in stop_words] for doc in texts]

data19 = data2019.description_processed.values.tolist()
data_words = list(sent_to_words(data19))
data_words = remove_stopwords(data_words)

print(data_words[:1][0][:30])
```

```python
import gensim.corpora as corpora
id2word = corpora.Dictionary(data_words)

texts= data_words

corpus = [id2word.doc2bow(text) for text in texts]

print(corpus[:1][0][:30])
```

```python
from pprint import pprint
num_topics = 10
lda_model = gensim.models.LdaMulticore(corpus=corpus,id2word=id2word,num_topics=num_topics)
pprint(lda_model.print_topic())
doc_lda = lda_model[corpus]
```
